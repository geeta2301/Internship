```python
# ICC-Men Team ranking
from bs4 import BeautifulSoup
import requests
```

```python
# Top 10 teams
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/t20i')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
TeamName=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    TeamName.append(i.text.split(','))
TeamName
```

```python
TeamName1=TeamName[0:10]
TeamName1
```

```python
matches=[]
for i in soup.find_all('td',class_='table-body__cell u-center-text'):
    matches.append(i.text)
matches
```

```python
matches1=[matches[i] for i in range(0,len(matches),2)][:9]
matches1
```

```python
matches1.insert(0,'52')
matches1
```

```python
points=[matches[i] for i in range(1,len(matches),2)][:9]  # points
points
```

```python
points.insert(0,'13,889')
points
```

```python
ratings=[ i.text for i in soup.find_all('td',class_='table-body__cell u-text-right rating')][:9]
ratings
```

```python
ratings.insert(0,'267')
ratings
```

```python
import pandas as pd
df=pd.DataFrame({'TeamName':TeamName1,'Matches':matches1, 'points':points, 'ratings':ratings})
df
```

```python
# Top 10 Batting players
page=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/t20i/batting')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Team=[]
for i in soup.find_all('span',class_="table-body__logo-text"):
    Team.append(i.text)
Team
```

```python
Team1=Team[0:9]
Team1.insert(0,'IND')
Team1
```

```python
Rating=[]
for i in soup.find_all('td',class_="table-body__cell rating"):
    Rating.append(i.text)
Rating
```

```python
Rating1=Rating[0:9]
Rating1.insert(0,'906')
Rating1
```

```python
Player=[]
for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    Player.append(i.text.split('\n')[1])
Player
```

```python
Player1=Player[0:9]
Player1.insert(0,'Suryakumar Yadav')
Player1
```

```python
df1=pd.DataFrame({'Player':Player1,'Team':Team1, 'Rating':Rating1})
df1
```
